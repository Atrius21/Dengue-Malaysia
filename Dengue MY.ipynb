## Data Visualization of Dengue cases in Malaysia 2010-2014

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('mohtraindata.csv')
df.head()
```

```python
df = df.replace('na', np.nan)
```

```python
df.isnull().sum()
```

df=df.dropna()
You can replace str 'na' with NaN or just fill it up with zeroes. 
Since we are doing data visualization only and not any predictions, we fill it up with zeroes instead.

```python
df.fillna(0,inplace=True)
df.head()
```

```python
df.columns = df.columns.str.lower()
df= df.drop(['malaysia'],axis=1)
df = df.rename(columns={'negeri':'week'})
df.head()
```

```python
df.week=df.week.map(lambda x: x.split(' ')[-1])
df.head()
```

```python
states = ['perlis', 'kedah', 'pulau pinang', 'perak', 'selangor',
       'wpkl/putrajaya', 'n.sembilan', 'melaka', 'johor', 'pahang',
       'terengganu', 'kelantan', 'sarawak', 'sabah', 'labuan']
```

```python
df[states] = df[states].astype(int)
df[states]= df[states]
df.head()
```

```python
df['week']=df['week'].astype(int)
df.info()
```

```python
dfsub = df.groupby(['year','week'])
dfsub.head(5)
```

```python
sns.set_style('white'),sns.set_style('ticks')
fig, (ax1,ax2) = plt.subplots(figsize=(15,10), nrows=2)
sns.barplot(data=df,x='week',y='selangor',hue='year',palette=sns.light_palette('Green'),
            order=df.week.iloc[:26],ax=ax1)
ax1.set_xlabel('Week',fontsize=15)
ax1.set_ylabel('Number of cases', fontsize=15)
ax1.set_title('Selangor State Dengue Cases', fontsize=20)


sns.barplot(data=df,x='week',y='selangor',hue='year',palette=sns.light_palette('Green'),
            order=df.week.iloc[26: 52],ax=ax2)
ax2.set(xlabel= 'Week', ylabel='Number of Cases', title='Selangor State Denggi Fever Case')
ax2.set_xlabel('Week',fontsize=15)
ax2.set_ylabel('Number of cases', fontsize=15)
ax2.set_title('Selangor State Dengue Cases', fontsize=20)

fig.tight_layout(h_pad=1.0)
plt.savefig('Selangorstate.png',format='png', dpi=600)
```
